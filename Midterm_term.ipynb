```python
# List of table names
dfs = [df14, df13, df12, df11, df10, df9, df8, df7, df6, df5, df4, df3, df2, df1]
concatenated_df = pd.concat(dfs).groupby('country').first().reset_index()

concatenated_df.columns
```

```python
concatenated_df
```

```python
selected_columns = ['country', 'prevalence', 'corruption_score', 'Average temprature', 'gini', 'rateWb', 'Avg size', 'Public services index', 'score', 'crime index', 'rank2022','latestRate', 'gendEqal2021', 'education', 'health','Water Scarcity', 'Air Quality' ]
```

```python
selected_columns
```

```python
# Use loc to select the columns by name and assign the result back to the DataFrame
df = concatenated_df.loc[:, selected_columns]
df
```

```python
df = df.rename(columns={'score': 'democracy_score', 'latestRate': 'litracy_rate', 'prevalence': 'depression_rate', 'rateWb': 'unemployment_rate', 'Avg size': 'family_avg_size', 'gendEqal2021': 'gender_equality', 'Average temprature': 'average_temperature', 'rank2022': 'peace_ranking'})
```

```python
df.loc[0, 'family_avg_size'] = 8.04
df.loc[0, 'gini'] = 59
df
```

```python
df['litracy_rate'] = df['litracy_rate'].round(2)
```

```python
#df['Water Scarcity'] = df['Water Scarcity'].astype(int)
print(df['Water Scarcity'].unique())
```

```python
print(df['Water Scarcity'].dtype)
```

```python
df['Water Scarcity'] = df['Water Scarcity'].astype(int)
```

```python
null_values = df.isnull().sum()
```

```python
null_values
```

```python
df = df.dropna(axis=0)
```

```python
df
```

```python
print(df['average_temperature'])
```

```python
df.dtypes
```

```python
# Remove the '°F' suffix and convert to float
df['average_temperature'] = df['average_temperature'].str.replace(' °F', '').astype(float)
```

```python
df
```

```python
# Convert Fahrenheit to Celsius
df['average_temperature'] = (df['average_temperature'] - 32) * 5/9
```

```python
#Round the values to two decimal places
df['average_temperature'] = df['average_temperature'].round(2)
```

```python
df.reset_index(drop = True)
```

```python
df.to_csv('clean_df.csv', index=False)
```
