```python
df_merge = df.merge(df_den, on='country')
df_merge
```

```python
merged_df = df_merge.merge(df_pov, on = 'country')
```

```python
merged_df 
```

```python
y = merged_df[['depression_rate']]
```

```python
X = merged_df[['corruption_score', 'average_temperature',
       'gini', 'unemployment_rate', 'Family size', 'Public services index',
       'crime index', 'litracy_rate', 'gender_equality', 'democracy_score',
       'Air Quality', 'Density (per sq km)', 'percentage_poverty',]]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=42)
```

```python

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
y_test_scaled = scaler.fit_transform(y_test)
y_train_scaled = scaler.fit_transform(y_test)
```

```python
plt.figure(figsize=(10, 6))
for i in range(X_scaled.shape[1]):
    plt.hist(X_scaled[:, i], bins=20, alpha=0.5, label='variables')
plt.xlabel('Scaled Values')
plt.ylabel('Frequency')
plt.title('Distribution of Scaled Features')
plt.legend()
plt.show()
```

```python
#pt = PowerTransformer()
#pt.fit(X_train)
#pt.fit(X_test)

#X_train_transformed = pt.transform(X_train)
#X_test_transformed = pt.transform(X_test)
```

```python
lm = LinearRegression()
model = lm.fit(X_train_scaled, y_train)
```

```python
print(model.coef_)
print(model.intercept_)
```

```python
y_pred = model.predict(X_test_scaled)
```

```python
y_test = np.array(y_test).ravel()
y_pred = np.array(y_pred).ravel()
compare = pd.DataFrame({'y_test':list(y_test), 'y_pred':list(y_pred)})
```

```python
import matplotlib.pyplot as plt

plt.scatter(y_test, y_pred)
plt.plot([min(y_test), max(y_test)], [min(y_pred), max(y_pred)], color='red', linestyle='--')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Scatter Plot of Actual vs. Predicted Values with Line')
plt.show()
```

```python
compare
```

```python
# Evaluate the model

mse = np.square(np.subtract(y_test,y_pred)).mean()
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred) 
```

```python
print('MSE: ', mse)
print('RMSE: ', rmse)
print('MAE: ', mae)
print('R2: ', r2)
```

```python
coefficients = model.coef_.flatten()
feature_importance = pd.DataFrame({'Features': X.columns, 'Coefficient': coefficients})
```

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Calculate the VIF for each feature
vif = pd.DataFrame()
vif["Variable"] = X.columns
vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

print(vif)
```

```python
import matplotlib.pyplot as plt
plt.scatter(vif['VIF'], vif['Variable'])
plt.xlabel('Variance Inflation Factor (VIF)')
plt.ylabel('Features')
plt.title('Scatter Plot of VIF values')
plt.show()
```
